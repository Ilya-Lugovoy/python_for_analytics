```python
penguins.dropna(inplace=True)
```

```python
sns.scatterplot(data=penguins, x='sex', y='body_mass_g')
```

```python
sns.scatterplot(data=penguins, x='flipper_length_mm', y='body_mass_g', hue='sex', style='island', size='species')
```

```python
sns.histplot(data=penguins, x='species')
```

```python
sns.histplot(data=penguins, x='sex')
```

```python
cols = penguins.columns
g = sns.PairGrid(penguins[cols])
g.map_diag(sns.histplot)
g.map_lower(sns.scatterplot)
g.map_upper(sns.scatterplot)
```

```python
cols = penguins.columns
sns.PairGrid(penguins[cols]).map(sns.scatterplot)
```

```python
penguins.corr()
```

```python
sns.heatmap(penguins.corr(), vmin=-1, vmax=1)
```

## Task 3

+ создать новый столбец в таблице с пингвинами, который будет отвечать за показатель длины клюва пингвина
 - high (от 42)
 - middle (от 35 до 42)
 - low (до 35)

```python
penguins.columns
```

```python
penguins[penguins['bill_length_mm'] < 50]
```

```python
penguins.loc[(penguins['bill_length_mm'] > 35) & (penguins['bill_length_mm'] < 42), 'new_col'] = 'middle'
```

```python
penguins.loc[penguins['bill_length_mm'] > 42, 'new_col'] = 'high'
```

```python
penguins = penguins.drop(columns=['new_col'])
```

```python
sns.histplot(data=penguins, x='flipper_length_mm', hue='new_col')
```
